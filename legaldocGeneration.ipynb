```python
from transformers import pipeline
import pandas as pd
from google.colab import files
import io
from transformers import AutoTokenizer, AutoModelForCausalLM # Import these as they are used in analyze_and_generate

# Load GPT-2 (These need to be loaded outside the function if the function is called multiple times)
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

# Use a generic sentiment analysis model
# Assuming sentiment_pipeline is also needed outside the function
sentiment_pipeline = pipeline("sentiment-analysis")


def analyze_and_generate(df: pd.DataFrame, text_column: str):
    """
    Analyzes sentiment and generates summaries for text in a specified DataFrame column.

    Args:
        df: The pandas DataFrame containing the text data.
        text_column: The name of the column containing the text to process.

    Returns:
        The DataFrame with 'Sentiment' and 'AI_Summary' columns added.
        Returns None if the specified text_column is not found.
    """
    sentiments = []
    summaries = []

    # Check if the specified text_column exists in the DataFrame
    if text_column not in df.columns:
        print(f"Error: Column '{text_column}' not found in the DataFrame.")
        print("Available columns:", df.columns.tolist())
        return None # Return None or raise an error if the column is not found

    for text in df[text_column]:
        # Ensure text is a string to avoid errors with the pipeline and tokenizer
        if not isinstance(text, str):
            text = str(text)
            # You might want to add a warning or log if you encounter non-string data

        try:
            sentiment = sentiment_pipeline(text)[0]['label']
        except Exception as e: # Catch a more general exception for the pipeline call
            print(f"Warning: Could not get sentiment for text '{text[:50]}...'. Error: {e}")
            sentiment = "N/A"

        prompt = f"Legal text: {text}\nSummary:"
        try:
            input_ids = tokenizer.encode(prompt, return_tensors="pt", truncation=True)
            output = model.generate(input_ids, max_length=60, pad_token_id=tokenizer.eos_token_id)
            summary = tokenizer.decode(output[0], skip_special_tokens=True)
        except Exception as e: # Catch exception for the generation process
             print(f"Warning: Could not generate summary for text '{text[:50]}...'. Error: {e}")
             summary = "Error generating summary"


        sentiments.append(sentiment)
        summaries.append(summary)

    df['Sentiment'] = sentiments
    df['AI_Summary'] = summaries
    return df

# Assuming df is already loaded from a previous cell.
# If not, you might need to include the file upload/loading code here.
# Example if df is NOT in scope:
# print("Please upload your CSV file.")
# uploaded = files.upload()
# if not uploaded:
#     print("No file uploaded. Exiting.")
# else:
#     file_name = list(uploaded.keys())[0]
#     df = pd.read_csv(io.BytesIO(uploaded[file_name]))
#     print(f"DataFrame loaded from '{file_name}'.")
#     print("DataFrame columns:", df.columns.tolist()) # Print columns after loading


# Call the function
# Make sure to replace 'legal_text' with the actual name of the column
# containing the text you want to analyze.
# Check the output of print("DataFrame columns:", df.columns.tolist()) above to confirm the name.
text_column_name = 'legal_text' # <--- REPLACE THIS WITH YOUR ACTUAL COLUMN NAME

if 'df' in globals(): # Check if df exists in the global scope
    results = analyze_and_generate(df.copy(), text_column_name) # Pass a copy to avoid modifying the original df directly if needed

    if results is not None: # Only proceed if the function returned a DataFrame
        print("\nAnalysis Complete. Results Head:")
        display(results.head()) # Use display for better formatting in Colab

        # Save and download the results
        output_filename = "analyzed_legal_documents.csv"
        results.to_csv(output_filename, index=False)
        print(f"\nResults saved to '{output_filename}'.")

        # Download the file
        try:
            files.download(output_filename)
            print(f"'{output_filename}' downloaded successfully.")
        except Exception as e:
            print(f"Error downloading file: {e}")
            print("You can manually download the file from the file explorer on the left.")
    else:
        print("\nAnalysis skipped due to missing text column.")

else:
    print("DataFrame 'df' not found. Please run the cells to load the data first.")
```

```python
!pip install gradio

import gradio as gr

def process_input(text):
    sentiment = sentiment_pipeline(text)[0]['label']
    prompt = f"Legal text: {text}\nSummary:"
    input_ids = tokenizer.encode(prompt, return_tensors="pt", truncation=True)
    output = model.generate(input_ids, max_length=60, pad_token_id=tokenizer.eos_token_id)
    summary = tokenizer.decode(output[0], skip_special_tokens=True)
    return sentiment, summary

gr.Interface(fn=process_input, inputs="text", outputs=["text", "text"],
             title="Legal Document Analyzer",
             description="Enter a legal sentence to get sentiment and AI summary").launch()
```
